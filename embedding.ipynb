```python
## first we are gonna do is to perform one hot encoding representation for a specific word
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### example sentence to check one_hot_encoding
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
##now We need to define the size of the vacab and convert each word into vectors
voc_size=10000
```

```python
### Performing one_hot Representation
one_hot_repr=[one_hot(words,voc_size)for words in sent]
## for every word in the sentence we are implementing one_hot thats why we have passed 2 parameters inside one_hot, voc_size will decide for each word it will create a matrix of size 10000(voc_size) and it will put 1 on the position where the word is present.
## And by covering it by square brackets we are creating a list of these vectors.
one_hot_repr
```

```python
## we can see the above o/p it 
##[6226, 5621, 4994, 8235]
##  the , glass, of, milk,  at 6226 position the is present so we will find one inside this matrix, and same goes for every word.


## Now we will take all these words sentence by sentence and pass them to word embedding layers and then we will convert it into word embedding representation


```

```python
### Word Embedding Representation  {Reffer notes from notebook}
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_repr, padding='pre',maxlen=sent_length)
print(embedded_docs)

## using pad_sequence we are making sixe of each vector same that will help training our model
```

```python
## Feature representation
dim=10 ## Dimension of our word embedding model
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')  ## adam is optimizer and mse = loss function
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
